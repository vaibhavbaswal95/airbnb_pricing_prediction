```python
import pandas as pd
```

```python
train = pd.read_csv('airbnb/train.csv')
```

```python
test = pd.read_csv('airbnb/test.csv')
```

```python
train['dataset'] = "train"
test['dataset'] = "test"
```

```python
train
```

```python
data = pd.concat([train,test],axis = 0)
```

```python
data.columns
```

```python
categorical = ['property_type','room_type','bed_type','cancellation_policy','city']
data = pd.get_dummies(data, columns = categorical)
```
